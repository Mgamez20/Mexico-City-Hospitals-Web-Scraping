```python
#Obtención de calificaciones y número de opiniones
def data_scraping(url, calif):
    browser.get(url)
    review_titles = browser.find_elements_by_class_name("jANrlb")
    
    for a in review_titles:
        calif.append(a.text)
       
    return calif
```

```python
import re

def lat_log(latitud,longitud,hospitales_url):
    
    for hospital in hospitales_url:
        geospacio = re.search(r"(?<=/@).*?(?=z/)", hospital).group(0)
        x = geospacio.split(',')
        latitud.append(x[0])
        longitud.append(x[1])
        
    return latitud,longitud
```

```python
latitud = []
longitud = []

latitud, longitud = lat_log(latitud, longitud, hospitales_url)
```
